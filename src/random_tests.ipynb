```python
import pandas as pd

crops_df = pd.read_csv("../Data/CleanCombinations/Combined_CR.csv")
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
crops_features_scaled = pd.DataFrame(scaler.fit_transform(crops_df[crops_df.columns[crops_df.columns != "label"]]))

crops_features_scaled
```

```python
crops_features_scaled["label"] = crops_df["label"]
crops_features_scaled
```

```python
crops_features_scaled.columns = crops_df.columns
crops_features_scaled
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(crops_df[crops_df.columns[crops_df.columns != "label"]], crops_df.label, random_state=1374)

x_train_s, x_test_s, y_train_s, y_test_s = train_test_split(crops_features_scaled[crops_features_scaled.columns[crops_features_scaled.columns != "label"]], crops_features_scaled.label, random_state=1374)
```

```python
y_test
```

```python
y_test_s
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier   

rfc = RandomForestClassifier(random_state=4731)
rfc_s = RandomForestClassifier(random_state=4731)

mlp = MLPClassifier(random_state=4731)
mlp_s = MLPClassifier(random_state=4731)

rfc.fit(x_train, y_train)
rfc_s.fit(x_train_s, y_train_s)

mlp.fit(x_train, y_train)
mlp.fit(x_train_s, y_train_s)
pass
```

```python
from sklearn.metrics import accuracy_score

y_pred = rfc.predict(x_test)
print(f"Normal: {accuracy_score(y_test, y_pred)}")

y_pred_s = rfc.predict(x_test_s)
print(f"Scaled: {accuracy_score(y_test_s, y_pred_s)}")
```

```python
y_pred = mlp.predict(x_test)
print(f"Normal: {accuracy_score(y_test, y_pred)}")

y_pred_s = mlp.predict(x_train_s)
print(f"Scaled: {accuracy_score(y_test_s, y_pred_s)}")
```
